```python
## filter out only recent items - appeared in stock in 2022 and 2023

def filter_data_by_date(data: dict) -> bool:
    date_text = data.get("details", {}).get("Date First Available", "")
    if not date_text:
        return True
    return int(date_text[-4:]) < 2022
```

```python
with gzip.open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics.jsonl.gz', 'rt') as fp:
    with open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics_2022_2023.jsonl','a',encoding='utf-8') as fp_out:
        with open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics_2022_2023_no_date.jsonl','a',encoding='utf-8') as fp_out_no_date:
            i =0
            for line in fp:
                data = json.loads(line.strip())
                try:
                    filter = filter_data_by_date(data)
                    if not filter:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                except:
                    json.dump(data, fp_out_no_date)
                    fp_out_no_date.write('\n')
                    fp_out_no_date.flush()
                i += 1
                if i % 10000 == 0:
                    print(f"Processed {i} lines")
        
```

```python
### split the items into two categories
## one file for items has main category information
## one file for items without main category information
def filter_category(data:dict)-> dict:
    filter = False
    if data['main_category'] == None:
        filter = True
    return filter
```

```python
with open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics_2022_2023.jsonl', 'rt', encoding='utf-8') as fp:
    with open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics_2022_2023_with_category.jsonl', 'a', encoding='utf-8') as fp_out:
        with open('D:/ai-engineering-bootcamp-prerequisites/data/meta_Electronics_2022_2023_no_category.jsonl', 'a', encoding='utf-8') as fp_out_no_category:
            i = 0
            for line in fp:
                data = json.loads(line.strip())
                if not filter_category(data):
                    json.dump(data, fp_out)
                    fp_out.write('\n')
                    fp_out.flush()
                else:
                    json.dump(data, fp_out_no_category)
                    fp_out_no_category.write('\n')
                    fp_out_no_category.flush()
                i += 1
                if i % 10000 == 0:
                    print(f"Processed {i} lines")
```
